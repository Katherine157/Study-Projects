Автоматизация работы секретаря

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

'''Пользователь по команде “p” может узнать владельца документа по его номеру'''

def owner_document():
  user_number = input('Введите номер документа: ') 
  for x in documents:
    if user_number == x['number']:
      return x['name']
  print('Документа с таким номером в базе нет')


'''Пользователь по команде “s” может по номеру документа узнать на какой полке он хранится'''

def shelf_number():
  doc_number = input('Введите номер документа: ') 
  for num in directories:
    if doc_number in directories[num]:
      return num
  print('Документа с таким номером на полках нет')


'''Пользователь по команде “l” может увидеть полную информацию по всем документам'''

def full_info_doc():
  for shelf, value in directories.items():
    for doc in documents:
      if doc['number'] in value:
        print(f"тип: {doc['type']}, номер: {doc['number']}, имя: {doc['name']}, полка хранения: {shelf}")


'''Пользователь по команде “ads” может добавить новую полку'''

def new_shelf():
  shelf = input('Введите номер новой полки: ')
  if shelf in directories:
    return f"Такая полка уже существует. Текущий перечень полок: {list(directories.keys())}"
  else:
    directories[shelf] = []
    return f"Полка добавлена. Текущий перечень полок: {list(directories.keys())}"


'''Пользователь по команде “ds” может удалить существующую полку из данных (только если она пустая)'''

def del_shelf():
  number_shelf = input('Введите номер полки для удаления: ')
  if number_shelf not in directories:
    print(f"Такой полки не существует. Текущий перечень полок: {list(directories.keys())}")
  elif directories[number_shelf]:
    print('На полке есть документы. Удаление не доступно.')
  else:
    del directories[number_shelf]
    print(f"Полка удалена. Текущий перечень полок: {list(directories.keys())}")



'''Пользователь по команде “ad” может добавить новый документ в данные'''

def add_doc():
  new_type = input('Введите тип документа: ')
  new_number = input('Введите номер документа: ')
  new_name = input('Введите владельца документа: ')
  shelf = input('Введите полку для хранения: ')

  if shelf in directories:
    value = {'type': new_type, 'number': new_number, 'name': new_name}
    documents.append(value)
    directories[shelf].append(new_number)
    return f"Документ создан. Текущий перечень документов: {list(documents)}"
  print(f"Полки с таким номером не существует. Добавьте полку через команду: ads. Текущий список полок: {list(directories.keys())}")


'''Пользователь по команде “d” может удалить документ из данных'''

def del_doc():
  del_num = input('Введите номер документа для удаления: ')
  for x in documents:
    if del_num == x['number']:
      documents.remove(x)
      return f"Документ удален. Текущий перечень документов: {list(documents)}"
  print(f"Документа с таким номером в базе нет. Текущий перечень документов: {list(documents)}")


'''Пользователь по команде “m” может переместить документ с полки на полку'''

def move_doc():
  doc_number = input('Введите номер документа: ')
  shelf_number = input('Введите номер полки: ')

  if shelf_number not in directories:
    return f"Полки с таким номером не существует. Добавьте полку через команду: ads. Текущий список полок: {list(directories.keys())}"
  
  for key, value in directories.items():
    if doc_number in value:
      directories[shelf_number] += [doc_number]
      value.remove(doc_number)
      return 'Документ успешно перемещен' 
  print('Документа с таким номером в базе нет')

'''Программа'''
''' q - прерывание действия программы'''

while True:
  print('Возможные команды: p, s, l, ads, ds, ad, d, m, q')
  comand = input('Введите команду: ')

  if comand == 'p':
    print(owner_document())

  if comand == 's':
    print(shelf_number())

  if comand == 'l':
    print(full_info_doc())

  if comand == 'ads':
    print(new_shelf())

  if comand == 'ds':
    print(del_shelf())

  if comand == 'ad':
    print(add_doc())

  if comand == 'd':
    print(del_doc())

  if comand == 'm':
    print(move_doc())    

  if comand == 'q':
    break  
    
```
